```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer,WordNetLemmatizer
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
text = "Natural Language processing enables machines to understand human languages"
tokens = word_tokenize(text)
```

```python
stop_words = set(stopwords.words('english')) # Corrected 'Step_words' to 'stop_words' and 'step_words' to 'stopwords'
filtered_tokens = [word for word in tokens if word.lower() not in stop_words] # Corrected 'step_words' to 'stop_words'
```

```python
ps = PorterStemmer()
stemmed_tokens = [ps.stem(word) for word in filtered_tokens]
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(word)for word in filtered_tokens] # Corrected 'lemmatizer_tokens' to 'lemmatized_tokens'
```

```python
print("Tokens:", tokens)
print("Filtered Tokens:", filtered_tokens)
print("Stemmed Tokens:", stemmed_tokens)
print("Lemmatized Tokens:", lemmatized_tokens)
```
